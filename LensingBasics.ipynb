```python
# imports all the modules needed

import numpy as np
import matplotlib.pyplot as plt
import pygravlens as gl
```

### Point Mass Model

```python
eins_pt = 0.606 # einstein_radius

plane_pt = gl.lensplane('ptmass', [0.0, 0.0, eins_pt])
model_pt = gl.lensmodel([plane_pt])
model_pt.info()
model_pt.tile()
model_pt.plot()

plane_pt; # objects 
model_pt; # objects
```

```python
# find beta to find source positions
theta_pos = 1.05 # arcsecs
theta_neg = -0.35 # arcsecs

betaOne = theta_pos - (eins_pt**2 / theta_pos)
# value of beta should be the same for when you use theta_neg to solve for beta.
'''
beta_neg = theta_neg - (eins_pt**2 / theta_neg)
'''
betaTwo = 0.0 # choose

betaArray = [betaOne, betaTwo] 

# use model.findimg and pass it the beta values to find image pos
# need to find source positions and give the model source positions to find images.
imgarr,muarr,tarr = model_pt.findimg(betaArray)
print('Here is the image array: \n', imgarr, '\n')
print('Here is the magnification array: \n', muarr, '\n')
print('Here is the time delay array: \n', tarr, '\n')
print('This was found for the following beta: \n', betaArray)
```

### SIS

```python
eins_iso = 0.7 # einstein_radius

plane_iso = gl.lensplane('SIS', [0.0, 0.0, eins_iso])
model_iso = gl.lensmodel([plane_iso])
model_iso.info()
model_iso.tile()
model_iso.plot(plotgrid=True,plotcrit='red')
```

```python
# find beta to find source positions for isothermal model
theta_pos = 1.05 # arcsecs
theta_neg = -0.35 # arcsecs

betaOne = theta_pos - eins_iso
betaTwo = 0.0 # choose

betaArray= [betaOne, betaTwo] 

# use model.findimg and pass it the beta values to find image pos
# need to find source positions and give the model source positions to find images.
imgarr,muarr,tarr = model_iso.findimg(betaArray)
print('Here is the image array: \n', imgarr, '\n')
print('Here is the magnification array: \n', muarr, '\n')
print('Here is the time delay array: \n', tarr, '\n')
print('This was found for the following beta: \n', betaArray)
```
